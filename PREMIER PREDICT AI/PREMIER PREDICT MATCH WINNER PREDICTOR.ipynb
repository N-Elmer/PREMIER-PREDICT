```python
#WE WANT TO CALL DATA VARIABLE FOR TABLE WITHOUT THE ABOVE REMOVED VALUES AS DATA1
data1 = data
```

```python
x = data1[["FTHG","FTAG","HTGS","ATGS","HTGC","ATGC","HTP","ATP","MW","HTFormPts","ATFormPts","HTWinStreak3","HTWinStreak5","HTLossStreak3","HTLossStreak5","ATWinStreak3","ATWinStreak5","ATLossStreak3","ATLossStreak5","HTGD","ATGD","DiffPts","DiffFormPts"]]
y = data1["FTR"]
```

```python
x = np.array(x)
y = np.array(y)
```

```python
#TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=None)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
#LIR = LinearRegression()
LOR = LogisticRegression()
#RFR = RandomForestRegressor()
KNN = KNeighborsClassifier(n_neighbors=3)
DTC = DecisionTreeClassifier()
SVM = SVC()
```

```python
# NOW WE WILL TRAIN OUR MODELS
#LItrain = LIR.fit(x_train, y_train)
LOtrain = LOR.fit(x_train, y_train)
#Rtrain = RFR.fit(x_train, y_train)
Ktrain = KNN.fit(x_train, y_train)
Dtrain = DTC.fit(x_train, y_train)
Strain = SVM.fit(x_train, y_train)
```

```python
# LIaccuracy = LIR.score(x_train, y_train)
# print("The Train Accuracy Percentage Is As Follows: \n",((LIaccuracy)*100),"\n")
```

```python
LOaccuracy = LOR.score(x_train, y_train)
print("The Train Accuracy Percentage Is As Follows: \n",((LOaccuracy)*100),"\n")
```

```python
# Raccuracy = RFR.score(x_train, y_train)
# print("The Train Accuracy Percentage Is As Follows: \n",((Raccuracy)*100),"\n")
```

```python
Kaccuracy = KNN.score(x_train, y_train)
print("The Train Accuracy Percentage Is As Follows: \n",((Kaccuracy)*100),"\n")
```

```python
Daccuracy = DTC.score(x_train, y_train)
print("The Train Accuracy Percentage Is As Follows: \n",((Daccuracy)*100),"\n")
```

```python
Saccuracy = SVM.score(x_train, y_train)
print("The Train Accuracy Percentage Is As Follows: \n",((Saccuracy)*100),"\n")
```

```python
# TEST OUR MODEL
#LItest = LItrain.predict(x_test)
#Rtest =  Rtrain.predict(x_test)
LOtest = LOtrain.predict(x_test)
Ktest = Ktrain.predict(x_test)
Dtest = Dtrain.predict(x_test)
Stest = Strain.predict(x_test)
```
